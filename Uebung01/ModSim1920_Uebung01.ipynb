Legen Sie ein Feld von x-Werten zwischen -pi und pi an, und plotten Sie `sin(x)*sin(5*x)` gegen `cos(x)*sin(5*x)`.

````python
x = []
i = -math.pi
while i <= math.pi:
    x.append(i)
    i += 0.01
y1 = [math.sin(i)*math.sin(5*i) for i in x]
y2 = [math.cos(i)*math.sin(5*i) for i in x]
plot(x,y1)
plot(x,y2)
````
